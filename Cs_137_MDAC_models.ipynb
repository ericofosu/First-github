```python
import numpy as np
```

```python
degree = 2
```

```python
print("For LaBr detector")
for nat in naturals:
  x = labr[nat]
  y = labr['Cs137mdac']
```

```python
coeffs = np.polyfit(x, y, degree)
poly = np.poly1d(coeffs)

y_pred = poly(x)
r2 = 1 -np.sum((y - y_pred)**2)/np.sum((y - np.mean(y))**2)
print("R2 score:", r2)

print(f"{nat} Polynomial coefficients:", coeffs, "\n")

plt.scatter(x, y, color='blue', label='Lanthanim Bromide detector Cs-137 MDAC against naturals')
plt.plot(np.sort(x), poly(np.sort(x)), color='red', label=f'Polynomial (deg{degree})')
plt.xlabel(f'{nat} (natural)')
plt.ylabel('Cs137mdac')
plt.title(f'LaBr: Cs137mdac vs {nat}\nR2 = {r2:.3f}')
plt.grid(True)
plt.legend()
plt.show()
```
