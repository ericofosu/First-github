```python
import numpy as np
```

```python
degree = 2
```

```python
print("For LaBr detector")
for nat in naturals:
  x = labr[nat]
  y = labr[' Co60mdac']
```

```python
coeffs = np.polyfit(x, y, degree)
poly = np.poly1d(coeffs)
```

```python
y_pred = poly(x)
r2 = 1 -np.sum((y - y_pred)**2)/np.sum((y - np.mean(y))**2)
print("R2 score:", r2)
```

```python
print(f"{nat} Polynomial coefficients:", coeffs, "\n")
```

```python
plt.scatter(x, y, color='black', label='Lanthanim Bromide detector Co-60 MDAC against naturals')
plt.plot(np.sort(x), poly(np.sort(x)), color='black', label=f'Polynomial (deg{degree})')
plt.xlabel(f'{nat} (natural)')
plt.ylabel(' Co60mdac')
plt.title(f'LaBr: Co60mdac vs {nat}\nR2 = {r2:.3f}')
plt.grid(True)
plt.legend()
plt.show()
```

```python
print("For SrI detector")
for nat in naturals:
  x = srl[nat]
  y = srl[' Co60mdac']
```

```python
coeffs = np.polyfit(x, y, degree)
poly = np.poly1d(coeffs)
```

```python
y_pred = poly(x)
r2 = 1 -np.sum((y - y_pred)**2)/np.sum((y - np.mean(y))**2)
print("R2 score:", r2)
```

```python
print(f"{nat} Polynomial coefficients:", coeffs, "\n")
```

```python
plt.scatter(x, y, color='black', label='Strontium Iodide Co-60 MDAC against naturals')
plt.plot(np.sort(x), poly(np.sort(x)), color='black', label=f'Polynomial (deg{degree})')
plt.xlabel(f'{nat} (natural)')
plt.ylabel(' Co60mdac')
plt.title(f'Srl: Co60mdac vs {nat}\nR2 = {r2:.3f}')
plt.grid(True)
plt.legend()
plt.show()
```
