```python
import pandas as pd
labr=pd.read_csv('Coefficients_80labr_30-31Gen2022.csv')
srl=pd.read_csv('Coefficients_90sri_30-31Gener2022_I131.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
naturals = ['BiPb214', 'Th232']
```

```python
for nat in naturals:
  plt.figure(figsize=(10,5))
  plt.scatter(labr[nat], labr['Cs137mdac'], color='blue', label='Lanthanim Bromide detector Cs-137 MDAC against naturals')
  plt.xlabel(f'{nat} (natural activity)')
  plt.ylabel('Cs137mdac')
  plt.title(f'Lanthanim Bromide detector Cs-137 MDAC against naturals')
  plt.grid(True)
  plt.show()
```

```python
for nat in naturals:
  plt.figure(figsize=(10,5))
  plt.scatter(srl[nat], srl['Cs137mdac'], color='red', label='Strontium Iodide detector Cs-137 MDAC against naturals')
  plt.xlabel(f'{nat} (natural activity)')
  plt.ylabel('Cs137mdac')
  plt.title(f'Strontium Iodide detector Cs-137 MDAC against naturals')
  plt.grid(True)
  plt.show()
```
